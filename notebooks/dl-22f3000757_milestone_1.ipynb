```python
stopwordss = set(stopwords.words('english'))
```

```python
# list(stopwordss)
```

```python
def remove_stopwords(text):
    words = text.split()
    words = [word for word in words if word not in stopwordss]
    return " ".join(words)
```

```python
df['text_no_stops'] = df['text_no_punct'].apply(remove_stopwords)
```
