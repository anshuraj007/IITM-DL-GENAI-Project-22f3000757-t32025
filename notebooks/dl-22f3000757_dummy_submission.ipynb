```python
train_df = pd.read_csv(TRAIN_CSV_PATH)
test_df = pd.read_csv(TEST_CSV_PATH)
```

```python
train_df.shape
```

```python
classes = ['anger','fear','joy','sadness','surprise']
```

```python
X_train_dummy = np.zeros((len(train_df), 1))
X_test_dummy = np.zeros((len(test_df), 1))
```

```python
from sklearn.dummy import DummyClassifier
```

```python
dummy_submission = test_df[['id']].copy()
```

```python
for c in classes:
    y_train = train_df[c].values
    dummy_classifier = DummyClassifier(strategy="uniform", random_state=42)
    dummy_classifier.fit(X_train_dummy, y_train)

    dummy_submission[c] = dummy_classifier.predict(X_test_dummy)
```

```python
dummy_submission.to_csv("submission.csv", index=False)
# print(dummy_submission.head())
```
